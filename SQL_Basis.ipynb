Q.9 What is the Difference Between a Primary Key and a Foreign Key?

...> 6ï¸âƒ£ Difference Between Primary Key and Foreign Key
Feature	Primary Key	Foreign Key
Purpose	Uniquely identifies a record	Links one table to another
Uniqueness	Must be unique	Can have duplicate values
NULL values	Not allowed	Allowed
Number per table	Only one	Multiple allowed
Reference	Identifies records in same table	Refers to primary key of another table
Example
-- Primary Key
employee_id INT PRIMARY KEY

-- Foreign Key
employee_id INT,
FOREIGN KEY (employee_id) REFERENCES employees(employee_id).
